```python
%matplotlib inline
import brfss
import pandas as pd

df = brfss.ReadBrfss()
df.head()
```

Using the data from the BRFSS, compute the linear least squares fit for log(weight) versus height. How would you best present the estimated parameters for a model like this where one of the variables is log-transformed? If you were trying to guess someoneâ€™s weight, how much would it help to know their height? 

```python
import thinkstats2
import numpy as np
import seaborn as sns
import thinkplot
import matplotlib.pyplot as plt

def least_squares(xs, ys):
    mean_x, var_x = xs.mean(), xs.var()
    mean_y = ys.mean()
    
    slope = thinkstats2.Cov(xs, ys, mean_x, mean_y) / var_x
    inter = mean_y - slope * mean_x
    
    return inter, slope 

df_clean = df.dropna(subset=['wtkg2', 'htm3'])
heights = df_clean.htm3
weights = np.log(df_clean.wtkg2)
least_squares(heights, weights)
```

```python
def sampling_distributions(live, iters=101):
    t = []
    for _ in range(iters):
        sample = thinkstats2.ResampleRows(live)
        heights = sample.htm3
        weights = np.log(sample.wtkg2)
        estimates = least_squares(heights, weights)
        t.append(estimates)
        
    inters, slopes = zip(*t)
    return inters, slopes

inters, slopes = sampling_distributions(df_clean)
```

```python
def plot_confidence_intervals(xs, inters, slopes, percent=90, **options):
    fys_seq = []
    for inter, slope in zip(inters, slopes):
        fxs, fys = thinkstats2.FitLine(xs, inter, slope)
        fys_seq.append(fys)
        
    p = (100 - percent) / 2
    percents = p, 100 - p
    low, high = thinkstats2.PercentileRows(fys_seq, percents)
    thinkplot.FillBetween(fxs, low, high, **options)

plot_confidence_intervals(heights, inters, slopes)
thinkplot.scatter(thinkstats2.Jitter(heights, 1), weights)
# plt.yscale('log')
thinkplot.show()
```
