```python
%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train_df = pd.read_csv("../datasets/train.csv")
test_df = pd.read_csv("../datasets/test.csv")

train_df.head()
```

```python
# Plots KDE for non-survivors
df_age = train_df[train_df.Survived == 0].dropna(subset=['Age'])
sns.kdeplot(df_age.Age, shade=True)

# Plots KDE for survivors
df_age = train_df[train_df.Survived == 1].dropna(subset=['Age'])
sns.kdeplot(df_age.Age, shade=True)

plt.show()
```

The above figure is a plot of the kernel density estimation, which estimates the probability density of a random variable. Essentially, this is a smoother representation of the histogram of a particular dataset. One interesting thing to note is that children have a fairly high survival rate, which makes sense since they would be the first people to get to the lifeboats.

```python
# Distribution based on fare
sns.violinplot(x=train_df.Survived, y=train_df.Fare)
plt.show()
```

Many people who did not pay much for their travel fare did not survive, and as the figure suggests, a good chunk of people who survived paid a lot of money to board the Titanic.

```python
# Distribution based on passenger class
sns.violinplot(x=train_df.Survived, y=train_df.Pclass)
plt.show()
```

Among the survivors, there was a pretty even distribution of people in different passenger classes.

```python
# Distribution based on port of embarkation
sns.violinplot(x=train_df.Survived, y=train_df.Embarked)
plt.show()
```

A lot of people who boarded at the Southampton port did not seem to survive the shipwreck.

```python
# Filters out people who are 13 years old or older
df_boys = train_df[(train_df.Age < 13) & (train_df.Sex == "male")]
df_girls = train_df[(train_df.Age < 13) & (train_df.Sex == "female")]

# Distribution based on age for boys
sns.kdeplot(df_boys[df_boys.Survived==0].Age, shade=True)
sns.kdeplot(df_boys[df_boys.Survived==1].Age, shade=True)
```

```python
# Distribution based on age for girls
sns.kdeplot(df_girls[df_girls.Survived==0].Age, shade=True)
sns.kdeplot(df_girls[df_girls.Survived==1].Age, shade=True)
```

From the above distributions, younger children had a higher probability of surviving, particularly young girls.
